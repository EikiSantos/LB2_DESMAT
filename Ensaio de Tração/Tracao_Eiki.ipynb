```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
amostra1 = pd.read_excel("tracao_1.xlsx")
amostra2 = pd.read_excel("tracao_2.xlsx")
amostra3 = pd.read_excel("tracao_3.xlsx")
```

```python
#Função que recebe uma serie e retorna uma lista com todos os valores da serie sem o primeiro argumento que é a unidade de medida:
def retorna_lista(serie):
    lista = []
    i = 1
    while i < len(serie):
        lista.append(serie[i])
        i += 1
    return lista

#Função que plota o histograma:
def hist(lista, n):
    plt.hist(lista, bins = n)
    plt.grid(True)
    plt.show();

def varre_lista(lista):
    saida = []
    i = 0
    while i < len(lista):
        if lista[i] <= 1:
            saida.append(lista[i])
        i += 1
    return saida

def reduct_lista(entrada, lista):
    saida = []
    i = 0
    while i < len(entrada):
        saida.append(lista[i])
        i += 1
    return saida
```

```python
#Amostra 1:
A1_forca = retorna_lista(amostra1.Força)
A1_deform = retorna_lista(amostra1["Deformação à tração (Extensômetro)"])
#Amostra 2:
A2_forca = retorna_lista(amostra2.Força)
A2_deform = retorna_lista(amostra2["Deformação à tração (Extensômetro)"])
#Amostra 3:
A3_forca = retorna_lista(amostra3.Força)
A3_deform = retorna_lista(amostra3["Deformação à tração (Extensômetro)"])

#Amostra 1 antes do 1% :
A1_deform_check = varre_lista(A1_deform)
A1_forca_check = reduct_lista(A1_deform_check, A1_forca)
#Amostra 2 antes do 1% :
A2_deform_check = varre_lista(A2_deform)
A2_forca_check = reduct_lista(A2_deform_check, A2_forca)
#Amostra 3 antes do 1% :
A3_deform_check = varre_lista(A3_deform)
A3_forca_check = reduct_lista(A3_deform_check, A3_forca)
```

```python
print("-"*55)
plt.plot(A1_forca_check, A1_deform_check,  color="orange")
plt.title("Corpo de Prova 1:")
plt.grid(True)
plt.xlabel("Tensão")
plt.ylabel("Deformação")
plt.show();
print("-"*55)
plt.plot(A2_forca_check, A2_deform_check,  color="cyan")
plt.title("Corpor de Prova 2:")
plt.grid(True)
plt.xlabel("Tensão")
plt.ylabel("Deformação")
plt.show();
print("-"*55)
plt.plot(A3_forca_check, A3_deform_check,  color="olive")
plt.title("Corpo de Prova 3:")
plt.grid(True)
plt.xlabel("Tensão")
plt.ylabel("Deformação")
plt.show();
print("-"*55)
```
