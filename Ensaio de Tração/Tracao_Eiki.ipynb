```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
#Função que recebe uma serie e retorna uma lista com todos os valores da serie sem o primeiro argumento que é a unidade de medida:
def retorna_lista(serie):
    lista = []
    i = 1
    while i < len(serie):
        lista.append(serie[i])
        i += 1
    return lista

#Função que plota o histograma:
def hist(lista, n):
    plt.hist(lista, bins = n)
    plt.grid(True)
    plt.show();

#Função que reduz os valores da deformação até 1% :
def varre_lista(lista):
    saida = []
    i = 0
    while i < len(lista):
        if lista[i] <= 1:
            saida.append(lista[i])
        i += 1
    return saida

#Função que iguala o tamanho da lista de forca e deformacao:
def reduct_lista(entrada, lista):
    saida = []
    i = 0
    while i < len(entrada):
        saida.append(lista[i])
        i += 1
    return saida

#Função que calcula a area da seção transversal do corpo de prova:
def calcula_area(diametro):
    area = math.pi*((diametro/2)**2)
    return area

#Função que retorna uma lista com a tensão:
def calcula_tensao(area, forca):
    saida = []
    i = 0
    while i < len(forca):
        saida.append(forca[i]/area)
        i += 1
    return saida

#Função que encontra a equação do primeiro grau:

def eq_afim(list_x, list_y):
    x1 = list_x[0]
    x2 = list_x[1]
    y1 = list_y[0]
    y2 = list_y[1]
    m = (y2 - y1)/(x2 - x1)
    k = y1 - m*x1
    return m, k 

#Função que retorna uma lista (tensao):
def return_list(m, k, inicio, fim, le):
    eixo_y = []
    if le == False:
        eixo_x = np.arange(inicio, fim, (fim-inicio)/len(A1_tensao))
    else:
        eixo_x = A3_deform_check
    i = 0
    while i < len(eixo_x):
        y = m*eixo_x[i] + k
        eixo_y.append(y)
        i += 1
    return eixo_y

def plota_eq_afim(m, k, inicio, fim):
    eixo_y = []
    eixo_x = np.arange(inicio, fim, 0.01)
    i = 0
    while i < len(eixo_x):
        y = m*eixo_x[i] + k
        eixo_y.append(y)
        i += 1
    plt.plot(eixo_x, eixo_y, color = "orange")
    
#Função que calcula o módulo de elasticidade:
def calcula_m_elast():

    A1_deform_elast = []
    for d in A1_deform_check:
        if 0.01 < d and d < 0.15:
            A1_deform_elast.append(d)

    A1_tensao_elast = []

    i = 0

    while i < len(A1_deform_elast):
        A1_tensao_elast.append(A1_tensao[i])
        i += 1
    x1 = A1_deform_elast[300]
    y1 = A1_tensao_elast[300]
    x2 = A1_deform_elast[2000]
    y2 = A1_tensao_elast[2000]
    m1 = (y1 - y2)/(x1 - x2)
#-------------------------------------------------------------------
    
    A2_deform_elast = []
    for d in A2_deform_check:
        if 0.01 < d and d < 0.11:
            A2_deform_elast.append(d)

    A2_tensao_elast = []

    i = 0

    while i < len(A2_deform_elast):
        A2_tensao_elast.append(A2_tensao[i])
        i += 1

    x1 = A2_deform_elast[300]
    y1 = A2_tensao_elast[300]
    x2 = A2_deform_elast[2000]
    y2 = A2_tensao_elast[2000]
    m2 = (y1 - y2)/(x1 - x2)
#-------------------------------------------------------------------

    A3_deform_elast = []
    for d in A3_deform_check:
        if 0.01 < d and d < 0.38:
            A3_deform_elast.append(d)

    A3_tensao_elast = []

    i = 0

    while i < len(A3_deform_elast):
        A3_tensao_elast.append(A3_tensao[i])
        i += 1

    x1 = A3_deform_elast[1900]
    y1 = A3_tensao_elast[1900]
    x2 = A3_deform_elast[4000]
    y2 = A3_tensao_elast[4000]
    m3 = (y1 - y2)/(x1 - x2)
    return [m1, m2, m3]

#Função que plota um gráfico com a linha de tendencia de uma eq afim:
def plota_2_graph(number, fator):
    A1_deform_elast = []
    for d in A1_deform_check:
        if 0.01 < d and d < 0.15:
            A1_deform_elast.append(d)

    A1_tensao_elast = []

    i = 0

    while i < len(A1_deform_elast):
        A1_tensao_elast.append(A1_tensao[i])
        i += 1
    x1 = A1_deform_elast[300]
    y1 = A1_tensao_elast[300]
    x2 = A1_deform_elast[2000]
    y2 = A1_tensao_elast[2000]
    m1 = (y1 - y2)/(x1 - x2)
#-------------------------------------------------------------------
    
    A2_deform_elast = []
    for d in A2_deform_check:
        if 0.01 < d and d < 0.11:
            A2_deform_elast.append(d)

    A2_tensao_elast = []

    i = 0

    while i < len(A2_deform_elast):
        A2_tensao_elast.append(A2_tensao[i])
        i += 1

    x1 = A2_deform_elast[300]
    y1 = A2_tensao_elast[300]
    x2 = A2_deform_elast[2000]
    y2 = A2_tensao_elast[2000]
    m2 = (y1 - y2)/(x1 - x2)
#-------------------------------------------------------------------

    A3_deform_elast = []
    for d in A3_deform_check:
        if 0.01 < d and d < 0.38:
            A3_deform_elast.append(d)

    A3_tensao_elast = []

    i = 0

    while i < len(A3_deform_elast):
        A3_tensao_elast.append(A3_tensao[i])
        i += 1

    x1 = A3_deform_elast[1900]
    y1 = A3_tensao_elast[1900]
    x2 = A3_deform_elast[4000]
    y2 = A3_tensao_elast[4000]
    m3 = (y1 - y2)/(x1 - x2)
    if number == 1 or number == 4:
        if fator == False:
            plt.plot(A1_deform_elast, A1_tensao_elast, color = "#855b21")
            m1, k1 = eq_afim([A1_deform_elast[300],A1_deform_elast[2000]],[A1_tensao_elast[300], A1_tensao_elast[2000]])
            plota_eq_afim(m1, k1, A1_deform_elast[300], A1_deform_elast[2000])
            plt.plot(A1_deform_elast[300], A1_tensao_elast[300], color = "#41b5a8", marker = "o")
            plt.plot(A1_deform_elast[2000], A1_tensao_elast[2000], color = "#41b5a8", marker = "o")
            plt.grid(True)
            plt.title("Corpo de Prova 1:")
            plt.xlabel("Deformação")
            plt.ylabel("Tensão")
            plt.show()
        elif fator == True:
            plt.plot(A1_deform_check, A1_tensao, color="brown")
            plota_eq_afim(m1, -0.02*m1, 0.02, A1_deform_check[4850])
            plt.grid(True)
            plt.title("Corpo de Prova 1:")
            plt.xlabel("Deformação")
            plt.ylabel("Tensão")
            plt.show()
        
            l1 = return_list(m1, -0.02*m1, 0.02, A1_deform_check[4850], False)
            i = 3000
            while i < len(A1_tensao):
                a = "%.4f" % l1[i]
                b = "%.4f" % A1_tensao[i]
                a = str(a)
                b = str(b)
                if a[2]==b[2] and a[3]==b[3]:
                    print("-"*55)
                    print("Limite de Escoamento: ", A1_tensao[i])
                    i = len(A1_tensao)-1
                i += 1

        print("-"*55)
        print("Coeficiente angular do Corpo de Prova 1: ", m1)
        print("-"*55)
    if number == 2 or number == 4:
        if fator == False:
            plt.plot(A2_deform_elast, A2_tensao_elast, color = "olive")
            m2, k2 = eq_afim([A2_deform_elast[300],A2_deform_elast[2000]],[A2_tensao_elast[300], A2_tensao_elast[2000]])
            plota_eq_afim(m2, k2, A2_deform_elast[300], A2_deform_elast[2000])
            plt.plot(A2_deform_elast[300], A2_tensao_elast[300], color = "#41b5a8", marker = "o")
            plt.plot(A2_deform_elast[2000], A2_tensao_elast[2000], color = "#41b5a8", marker = "o")
            plt.grid(True)
            plt.title("Corpo de Prova 2:")
            plt.xlabel("Deformação")
            plt.ylabel("Tensão")
            plt.show()
        elif fator == True:
            plt.plot(A2_deform_check, A2_tensao, color="brown")
            plota_eq_afim(m2, -0.02*m2, 0.02, A2_deform_check[4400])
            plt.grid(True)
            plt.title("Corpo de Prova 2:")
            plt.xlabel("Deformação")
            plt.ylabel("Tensão")
            plt.show()
            l2 = return_list(m2, -0.02*m2, 0.02, A2_deform_check[4400], False)
            i = 3000
            while i < len(A2_tensao):
                a = "%.4f" % l2[i]
                b = "%.4f" % A2_tensao[i]
                a = str(a)
                b = str(b)
                if a[2]==b[2] and a[3]==b[3]:
                    print("-"*55)
                    print("Limite de Escoamento: ", A2_tensao[i])
                    i = len(A2_tensao)-1
                i += 1
        
        
        print("-"*55)
        print("Coeficiente angular do Corpo de Prova 2: ", m2)
        print("-"*55)
    if number == 3 or number == 4:
        if fator == False:
            plt.plot(A3_deform_elast, A3_tensao_elast, color = "orange")
            m3, k3 = eq_afim([A3_deform_elast[1900],A3_deform_elast[4000]],[A3_tensao_elast[1900], A3_tensao_elast[4000]])
            plota_eq_afim(m3, k3, A3_deform_elast[1900], A3_deform_elast[4000])
            plt.plot(A3_deform_elast[1900], A3_tensao_elast[1900], color = "#41b5a8", marker = "o")
            plt.plot(A3_deform_elast[4000], A3_tensao_elast[4000], color = "#41b5a8", marker = "o")
            plt.grid(True)
            plt.title("Corpo de Prova 3:")
            plt.xlabel("Deformação")
            plt.ylabel("Tensão")
            plt.show()
        elif fator == True:
            plt.plot(A3_deform_check, A3_tensao, color="brown")
            plota_eq_afim(m3, -0.02*m3, 0.02, A3_deform_check[6700])
            plt.grid(True)
            plt.title("Corpo de Prova 3:")
            plt.xlabel("Deformação")
            plt.ylabel("Tensão")
            l3 = return_list(m3, -0.02*m3, 0.02, A3_deform_check[6699], True)
            plt.show()
            i = 1000
            while i < len(A3_tensao):
                a = "%.4f" % l3[i]
                b = "%.4f" % A3_tensao[i]
                a = str(a)
                b = str(b)
                if a[2]==b[2] and a[3]==b[3]:
                    print("-"*55)
                    print("Limite de Escoamento: ", A3_tensao[i])
                    i = len(A3_tensao)-1
                i += 1
        
        print("-"*55)
        print("Coeficiente angular do Corpo de Prova 3: ", m3)
        print("-"*55)
        
        
```
