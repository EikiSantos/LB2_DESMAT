```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
amostra1 = pd.read_excel("tracao_1.xlsx")
amostra2 = pd.read_excel("tracao_2.xlsx")
amostra3 = pd.read_excel("tracao_3.xlsx")
```

```python
#Função que recebe uma serie e retorna uma lista com todos os valores da serie sem o primeiro argumento que é a unidade de medida:
def retorna_lista(serie):
    lista = []
    i = 1
    while i < len(serie):
        lista.append(serie[i])
        i += 1
    return lista

#Função que plota o histograma:
def hist(lista, n):
    plt.hist(lista, bins = n)
    plt.grid(True)
    plt.show();

#Função que reduz os valores da deformação até 1% :
def varre_lista(lista):
    saida = []
    i = 0
    while i < len(lista):
        if lista[i] <= 1:
            saida.append(lista[i])
        i += 1
    return saida

#Função que iguala o tamanho da lista de forca e deformacao:
def reduct_lista(entrada, lista):
    saida = []
    i = 0
    while i < len(entrada):
        saida.append(lista[i])
        i += 1
    return saida

#Função que calcula a area da seção transversal do corpo de prova:
def calcula_area(diametro):
    area = math.pi*((diametro/2)**2)
    return area

#Função que retorna uma lista com a tensão:
def calcula_tensao(area, forca):
    saida = []
    i = 0
    while i < len(forca):
        saida.append(forca[i]/area)
        i += 1
    return saida
```

```python
#Amostra 1:
A1_forca = retorna_lista(amostra1.Força)
A1_deform = retorna_lista(amostra1["Deformação à tração (Extensômetro)"])
A1_deform_check = varre_lista(A1_deform)
A1_forca_check = reduct_lista(A1_deform_check, A1_forca)
A1_area = calcula_area(8.5)
A1_tensao = calcula_tensao(A1_area, A1_forca_check)

#Amostra 2:
A2_forca = retorna_lista(amostra2.Força)
A2_deform = retorna_lista(amostra2["Deformação à tração (Extensômetro)"])
A2_deform_check = varre_lista(A2_deform)
A2_forca_check = reduct_lista(A2_deform_check, A2_forca)
A2_area = calcula_area(8.3)
A2_tensao = calcula_tensao(A2_area, A2_forca_check)

#Amostra 3:
A3_forca = retorna_lista(amostra3.Força)
A3_deform = retorna_lista(amostra3["Deformação à tração (Extensômetro)"])
A3_deform_check = varre_lista(A3_deform)
A3_forca_check = reduct_lista(A3_deform_check, A3_forca)
A3_area = calcula_area(8.4)
A3_tensao = calcula_tensao(A3_area, A3_forca_check)
```

```python
def plota_graph_tracao():
    plt.plot(A1_deform_check, A1_tensao, color="brown")
    plt.grid(True)
    plt.title("Corpo de Prova 1:")
    plt.xlabel("Deformação")
    plt.ylabel("Tensão")
    plt.show()
    print("-"*55)
    plt.plot(A2_deform_check, A2_tensao, color="olive")
    plt.grid(True)
    plt.title("Corpo de Prova 2:")
    plt.xlabel("Deformação")
    plt.ylabel("Tensão")
    plt.show()
    print("-"*55)
    plt.plot(A3_deform_check, A3_tensao, color="orange")
    plt.grid(True)
    plt.title("Corpo de Prova 3:")
    plt.xlabel("Deformação")
    plt.ylabel("Tensão")
    plt.show()
    print("-"*55)
```

```python
plota_graph_tracao()
```
